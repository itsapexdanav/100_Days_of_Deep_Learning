# 🧠 What is LeNet-5? <br>

LeNet-5 is a classic CNN by Yann LeCun used for digit recognition (like MNIST).

# Layers: <br>
Conv1 → 6 filters (5×5) → ReLU <br>
Pooling1 → AvgPool (2×2) <br>
Conv2 → 16 filters (5×5) → ReLU <br>
Pooling2 → AvgPool (2×2)<br>
## Flatten
FC1 → 120 neurons<br>
FC2 → 84 neurons<br>
Output → 10 classes (Softmax)<br>

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize
x_train, x_test = x_train / 255.0, x_test / 255.0

# Reshape to (28,28,1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Build LeNet model
model = models.Sequential()

# C1
model.add(layers.Conv2D(6, (5,5), activation='relu', input_shape=(28,28,1)))

# S2
model.add(layers.AveragePooling2D(pool_size=(2,2)))

# C3
model.add(layers.Conv2D(16, (5,5), activation='relu'))

# S4
model.add(layers.AveragePooling2D(pool_size=(2,2)))

# Flatten
model.add(layers.Flatten())

# FC layers
model.add(layers.Dense(120, activation='relu'))
model.add(layers.Dense(84, activation='relu'))

# Output
model.add(layers.Dense(10, activation='softmax'))

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(x_train, y_train, epochs=5, validation_split=0.1)

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)

print("Test Accuracy:", test_acc)
```
