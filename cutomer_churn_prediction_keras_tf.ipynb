```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
# making layers
#1. Hidden_leayer with 3 neurons
model.add(Dense(3,activation='sigmoid',input_dim=11))
# “A Dense layer is a fully connected neural network layer where each neuron receives input from all neurons in the previous layer.”

#2.Output layer
model.add(Dense(1,activation='sigmoid'))


```

```python
model.summary()
```

```python
# model compilation
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(X_train_scaled,y_train,epochs=10)
```

```python
# weight and biases values
model.layers[0].get_weights()
```

```python
# model [rediction with probabilities futher we have to set threshold and convert to 0 or 1]
y_log=model.predict(X_test_scaled)
```

```python
#threshold
import numpy as np
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
